````python
# Example prompt that asks the LLM to return JSON data
prompt = f"""
            read the json file from {problems}, there are 25 problems from Problem 1 to Problem 25. 
            look at the problems one by one, find the final answer for each problem from five solutions in solutions field.
            and then compare the answers with the value in the answer_value field for each problem to see how many of them are correct.
            if all 5 answers are correct, output correct_final_answer: 5/5, pay attention to laTex format answer and mixed fraction.
            Ensure the output is ONLY the JSON object, nothing else. remove json from the beginning of the output.
                        Only output raw JSON with no extra text

"""

response = openai.ChatCompletion.create(
    model=model_name,  # or "gpt-3.5-turbo"
    messages=[
        {"role": "user", "content": prompt}
    ],
    temperature=0.2
)

# Extract JSON string from response (assumes the response is raw JSON)
content = response["choices"][0]["message"]["content"]

# Optional: Clean the output if wrapped in triple backticks or quotes
def clean_llm_json_output(raw_text):
    raw_text = raw_text.strip()
    if raw_text.startswith("```json"):
        raw_text = raw_text[7:]  # remove ```json
    elif raw_text.startswith("''' json"):
        raw_text = raw_text[8:]  # remove ''' json
    if raw_text.endswith("```") or raw_text.endswith("'''"):
        raw_text = raw_text[:-3]
    return raw_text.strip()

cleaned_json_str = clean_llm_json_output(content)

# Parse and save to file
try:
    data = json.loads(cleaned_json_str)
    with open("generated_users.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)
    print("✅ JSON saved to generated_users.json")
except json.JSONDecodeError as e:
    print("❌ Failed to parse JSON:", e)
    print("Raw LLM output:\n", content)
````
